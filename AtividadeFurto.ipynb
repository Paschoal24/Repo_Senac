```python
import pandas as pd 
```

```python
df = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
df
```

```python
furtos_por_delegacia = df.groupby('cisp')['furto_celular'].sum().reset_index()
furtos_por_delegacia.columns = ['delegacia', 'total_furtos']
```

```python
media = furtos_por_delegacia["total_furtos"].mean()
mediana = furtos_por_delegacia["total_furtos"].median()
q1 = furtos_por_delegacia["total_furtos"].quantile(0.25)
q3 = furtos_por_delegacia["total_furtos"].quantile(0.75)

# Exibir estatísticas
estatisticas = pd.DataFrame({
    "Estatística": ["Média", "Mediana", "1º Quartil (Q1)", "3º Quartil (Q3)"],
    "Valor": [media, mediana, q1, q3]
})
estatisticas
```

```python
delegacias_menos_furtos = furtos_por_delegacia[furtos_por_delegacia["total_furtos"] <= q1]
delegacias_menos_furtos
```

```python
delegacias_mais_furtos = furtos_por_delegacia[furtos_por_delegacia["total_furtos"] >= q3]
delegacias_mais_furtos
```
