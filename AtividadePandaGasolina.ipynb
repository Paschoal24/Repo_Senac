```python
import pandas as pd
```

```python
df_combustiveis = pd.read_csv('Preços semestrais -AUTOMOTIVOS_2025.01.csv', sep=';', encoding= 'latin1')
df_combustiveis
```

```python
df_rj = df_combustiveis.loc[df_combustiveis['Municipio'] == 'RIO DE JANEIRO']
df_rj
```

```python
df_combustiveis['Valor de Venda'] = df_combustiveis['Valor de Venda'].str.replace(',','.').astype(float)
df_combustiveis
```

```python
df_etanol = df_combustiveis[df_combustiveis['Produto'] == 'ETANOL']
df_etanol
```

```python
df_etanol_rj = df_etanol.groupby('Bairro')['Valor de Venda'].mean()
df_etanol_rj = df_etanol_rj.sort_values()
df_etanol_rj
```

```python
df_etanol_rj = df_etanol.groupby('Bairro')['Valor de Venda'].mean()
df_etanol_rj = df_etanol_rj.sort_values(ascending=False)
df_etanol_rj
```
